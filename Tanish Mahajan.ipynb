# Q2:- Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

```python
cd=data.query("location=='Argentina-CABA' & unit=='cases'")
cd1=data.query("location=='Argentina-Catamarca' & unit=='cases'")
index=cd['report_date'].unique()


cd[['value']]=(pd.to_numeric(cd.value))
cd.drop(['time_period','time_period_type'],axis=1,inplace=True)
a=cd.groupby("report_date").sum()
a=a['value'].tolist()



cd1[['value']]=(pd.to_numeric(cd.value))
cd1.drop(['time_period','time_period_type'],axis=1,inplace=True)
a1=cd1.groupby("report_date").sum()
a1=a1['value'].tolist()

df = pd.DataFrame({'Argentina-CABA': a,'Argentina-Catamarca': a1},index=index)
ax=df.plot.line(rot=0)
plt.xticks(rotation='45')
```

# Q3:- Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

```python
cd=data.query("location=='Argentina-CABA'")
cd[['value']]=(pd.to_numeric(cd.value))
cd.drop(['time_period','time_period_type'],axis=1,inplace=True)
a=cd.groupby("data_field").sum()
a.plot.pie(subplots=True,autopct='%1.1f%%',figsize=(5,5),labeldistance=1,pctdistance=2.5)
plt.title("Argentina-CABA")
plt.legend(loc=(1.9,0.5))




cd=data.query("location=='Argentina-Buenos_Aires'")
cd[['value']]=(pd.to_numeric(cd.value))
cd.drop(['time_period','time_period_type'],axis=1,inplace=True)
a=cd.groupby("data_field").sum()
a.plot.pie(subplots=True,autopct='%1.1f%%',figsize=(6,5),labeldistance=1.8,pctdistance=1.6)
plt.title("Argentina-Buenos_Aires")
plt.legend(loc=(1.9,1))



cd=data.query("location=='Argentina-Catamarca'")
cd[['value']]=(pd.to_numeric(cd.value))
cd.drop(['time_period','time_period_type'],axis=1,inplace=True)
a=cd.groupby("data_field").sum()
a.plot.pie(subplots=True,autopct='%1.1f%%',figsize=(5,5),labeldistance=1.5,pctdistance=1)
plt.title("Argentina-Catamarca")
plt.legend(loc=(1.5,0.5))
```
